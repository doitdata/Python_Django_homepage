# unit02,3 서울의 기온 데이터 분석하기

```python
import csv 
#seoul.csv 파일을 읽기 모드로 읽어오되 cp949(윈도우 한글 인코딩 방식)으로 읽어오라는 의미
#출력된 데이터의 특징 => 리스트/문자열/누락된 데이터 여부
f = open('D:/Python_Project/data_sample/Unit02/seoul.csv','r',encoding='cp949')
#읽어온 파일은 delimiter는 구분자라는 의미로 콤마 기준으로 분리해라 라는 의미
data= csv.reader(f, delimiter=',')
# header는 데이터의 속성을 설명, header를 별도로 저장하려면 next 함수를 사용하면 됨 
header = next(data)
#print(header)
max_temp = -999 #최고 기온을 저장할 변수 저장하기
max_date = ''
for row in data:
    if row[-1] == '': #-1인 이유는 공백값이 -1 (최고기온) 위치에 있기 때문
        row[-1]=-999 #공백값은 -999를 넣어 최고기온을 -999로 저장
    row[-1]=float(row[-1]) # 숫자로 다루기 위해서 숫자데이터로 변환. 실수형이므로 float 사용
    if max_temp < row[-1]: # 만약 지금까지 최고 온도보다 더 높으면 업데이트
        max_date = row[0]
        max_temp = row[-1]
        
f.close() #파일 닫기
print('기상 관측 이래 서울의 최고 기온이 가장 높았던 날은', max_date  +'로', max_temp, '도 였습니다.' )
```

```python
#데이터를 월별로 
import csv
import matplotlib.pyplot as plt

f =open('d:/Python_Project/data_sample/Unit06/seoul.csv')
data = csv.reader(f)
next(data)

#월별 데이터를 저장할 리스트 12개 생성
month = [[],[],[],[],[],[],[],[],[],[],[],[]]

for row in data:
    if row[-1] != '' :
        #월과 같은 번호이 인덱스에 월별데이터 저장(1월 => month[0])
        month[int(row[0].split('-')[1])-1].append(float(row[-1]))
        
        
plt.boxplot(month)
plt.show()
#여기서 동그라미는 이상치로 너무 크거나 작은값을 의미
```

```python
#8월 일별 기온데이터를 상자그림으로 표현하기
import matplotlib.pyplot as plt
import csv

f =open('d:/Python_Project/data_sample/Unit06/seoul.csv')
data = csv.reader(f)
next(data)

day=[]
for i in range(31):
    day.append([]) #day리스트 내에 31개 리스트 생성

for row in data:
    if row[-1] != '' :
        if row[0].split('-')[1] =='08' :
            day[int(row[0].split('-')[2])-1].append(float(row[-2]))
            
plt.style.use('ggplot') # 그래프 스타일 지정
plt.figure(figsize=(10,5), dpi=300) #그래프 크기수 지정
plt.boxplot(day, showfliers=False) #아웃라이어(이상치) 값 생략

plt.show()
```
