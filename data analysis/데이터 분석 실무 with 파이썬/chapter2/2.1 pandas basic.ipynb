```python
code_master = pd.read_excel('./files/sample_codemaster.xlsx')
code_master
```

```python
sample_1_code = pd.merge(
                left=sample_1, # left는 왼쪽 테이블을 sample_1로 설정
                right = code_master, # right는 오른쪽 테이블을 code_mastrt로 설정
                how='left', # how는 왼쪽 테이블을 기준으로 두 테이블을 결합, 왼쪽기준으로 값이 오른쪽에 없으면 NAN 표시
                left_on = '국적코드', # left_on은 왼쪽 테이블의 기준 칼럼은 국적코드
                right_on = '국적코드') # right_on은 오른쪽 테이블의 기준 칼럼은 국적코드
sample_1_code
```

```python
sample_1_code_inner = pd.merge(
        left=sample_1, # left는 왼쪽 테이블을 sample_1로 설정
        right=code_master, # right는 오른쪽 테이블을 code_mastrt로 설정
        how='inner', # inner는 두 테이블의 기준 칼럼의 값이 서로 일치하는 경우에만 통합하는 옵셥 
        left_on='국적코드', # left_on은 왼쪽 테이블의 기준 칼럼은 국적코드
        right_on='국적코드' # right_on은 오른쪽 테이블의 기준 칼럼은 국적코드
)
sample_1_code_inner
```

```python
sample_2 = pd.read_excel('./files/sample_2.xlsx',
                        header = 1,
                        skipfooter=2,
                        usecols='A:C')
sample_2['기준연월'] = '2019-2'
sample_2_code = pd.merge(left=sample_2,
                        right=code_master,
                        how='left',
                        left_on = '국적코드',
                        right_on = '국적코드')
sample_2_code
```

```python
# pivot_table() 메소드는 sample 데이터를 피벗화
# pivot_table() 함수의 aggfunc 옵션
# mean 평균 / sum 합계 / min 최소값 / max 최대값 / median 중앙값 / count 개수 / nunique 중복을 제거한 후 개수 (원소개수)  

sample_pivot = sample.pivot_table(values='입국객수', # 피벗테이블의 내부에 들어갈 값은 입국 객수
                                 index= '국적명', # 피벗테이블의 행(row)에 국적명에 배치
                                 columns = '기준년월',# 피벗테이블의 열(column)에 기준년월을 배치
                                 aggfunc='mean')# value를 인덱스와 컴럼 기준으로 평균을 계산
sample_pivot
```

```python
sample_pivot2 = sample.pivot_table(values='입국객수',
                                 index= '국적명',
                                 aggfunc='mean')
sample_pivot2
```
