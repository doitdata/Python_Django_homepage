### 157번 현재 시간을 년-월-일 시:분:초로 출력하기 localtime, strtime

time 모듈의 localtime()은 대한민국의 현재시간을 time.struct_time형식의 데이터로 리턴

#### strucr_time 형식은 다음의 9개의 속성을 가지고 있으며, 각 값들은 속이나 인덱스로 접근이 가능
#### 예를 들어 t = localtime()으로 현재 시간을 구해 변수 t에 지정했을 때, 현재 년도를 구하려면 t[0] 또는 t.tm_year로 접급
0 : tm_year : 현재 년도  
1 : tm_mon  : 현재 월  
2 : tm_mday :  현재 날짜  
3 : tm_hour : 현재 시간  
4 : tm_min : 현재 분  
5 : tm_sec : 현재 초    
6 : tm_wday : 현재 요일    
7 : tm_yday : 1월 1일부터 현재까지 날짜 수   
8 : tm_isdst : 섬머 타임일 경우 1, 아닐 경우 0, 모를 경루 1

#### strftime()은 struct_time 형식의 데이터를 포맷문자열을 이용해 특정 형식으로 시간을 표현해줌  
%Y : 년도를 나타냄   
%m : 월  
%d : 일  
%H : 시간을 24시간 표시 형식으로 나타냄   
%M : 분을 나타냄  
%S : 초를 나타냄  
%x : 현재 날짜를 월/일/년으로 나타냄  
%X : 현재 시간을 시:분:초로 나타냄 

```python
from time import localtime, strftime

logfile = 'test.log'

def writelog(logfile, log):
    time_stamp = strftime('%Y-%m-%d %X\t', localtime())
    log = time_stamp + log +'\n'
    
    with open(logfile, 'a') as f:
        f.writelines(log)
        
writelog(logfile, '2번째 로깅 문장입니다')
```

### 165번 스택 구현하기

스택은 나중에 저장된 자료가 먼저 추출될 수 있도록 되어있는 나열구조

```python
mystack = []

def putdata(data):
    global mystack
    mystack.append(data)
    
def popdata():
    global mystack
    if len(mystack) == 0:
        return None
    return mystack.pop()


putdata('데이터1')
putdata([3,4,5,6])
putdata(12345)

print('<스택상태>: ', end=''); print(mystack)

ret = popdata()
while ret != None:
    print('스택에서 데이터 추출: ', end = ''); print(ret)
    print('<스택상태>: ', end = ''); print(mystack)
    ret = popdata()
```

### 166번 문장에 나타나는 문자 빈도수 계산하기

```python
def getTextFreq(filename):
    with open(filename, 'r') as f:
        text = f.read()
        fa = {}
        for c in text:
            if c in fa:
                fa[c] += 1
            else:
                fa[c] = 1
    return fa

ret = getTextFreq('./mydata.txt')
ret = sorted(ret.items(), key = lambda x:x[1], reverse = True) # 내림차순으로 정렬

for c, freq in ret:
    if c == '\n':
        continue
    
    print('[%c] -> [%d]회 나타남' % (c, freq))
```

### 168번 파일에서 특정단어 개수세기

```python
def countWord(filename, word):
    with open(filename, 'r') as f:
        text =f.read()
        text = text.lower()
        pos = text.find(word) # word가 최초로 나타나는 위치를 반환
        count =0
        while pos != -1: # find가 찾는 문자열이 없는 경우 -1 반환
            count += 1
            pos = text.find(word, pos +1 )
            
    return count

word = input('mydata.txt에서 개수를 구할 단어를 입력하세요: ')
word = word.lower()
ret = countWord('mydata.txt', word)
print('[%s]의 개수: %d' %(word, ret))
```

### 177번 디렉터리를 하나의 ZIP 압출 파일로 만들기

```python
os.walk는 주어진 디렉터리를 모두 탐색하는 모듈
```

```python
from zipfile import *
import os

def compressAll(zipname, folder):
    print('[%s] -> [%s] 압축' % (folder, zipname))
    with ZipFile(zipname, 'w') as ziph:
        for dirname, subdirs, files in os.walk(folder):
            for file in files:
                ziph.write(os.path.join(dirname, file))
                
folder = 'tmp'
zipname = folder + '.zip'
compressAll(zipname, folder)
```

```python
from zipfile import *

def extractZip(zipname):
    with ZipFile(zipname, 'r') as ziph:
        ziph.extractall()
        print('[%s]가 성공적으로 추출되었습니다' % zipname)
        
extractZip('tmp.zip')
```

### 179번 로또 번호 추출기 만들기

```python
from random import shuffle
from time import sleep

gamenum = input('로또 게임 회수를 입력하세요: ')

for i in range(int(gamenum)):
    balls = [x+1 for x in range(45)]
    ret = []
    for j in range(6):
        shuffle(balls)
        number = balls.pop()
        ret.append(number)
    ret.sort()
    print('로또번호[%d]: ' %(i+1), end='')
    print(ret)
    sleep(1)
```

### 180번 남녀 파트너 정해주기 프록그램 만들기

```python
from random import shuffle

male = ['슈퍼맨', '심봉사', '아이언맨', '엔트맨']
female = ['원더우먼', '뺑덕', '줄리엣' '춘향이']

shuffle(male)
shuffle(female)
couples = zip(male, female)

for i, couple in enumerate(couples):
    print('커플 [%s] - [%s]'  %(couple[0], couple[1]))
```
