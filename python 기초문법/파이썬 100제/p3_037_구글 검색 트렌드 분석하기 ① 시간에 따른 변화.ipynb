```python
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
import os
```

```python
# 검색 keyword, 검색기간 입력
keyword = 'Galaxy Fold'
period = 'today 3-m' #검색기간, 최근 3개월

# Google Trend 접속
# pytrends.request 모듈의 TrenReq()클래스를 사용하여 검색 객체 TrendReq를 생성
trend_obj = TrendReq()
# 검색 키워드와 검색 기간을 build_pytrend() 메소드에 입력하여 trend_obj 객체에 전달
trend_obj.build_payload(kw_list=[keyword], timeframe=period)

# 시간에 따른 Trend 변화
# 시간에 따른 검색 트렌드 데이터를 찾기 위해 interest_over_time() 메소드를 이용해 검색결과를 데이터 프레임에 저장
trend_df = trend_obj.interest_over_time()
print(trend_df.head())

# 그래프 출력
# matplotlib 라이브러리의 스타일 서식을 지정하는 명령, ggplot 이라는 스타일을 적용
plt.style.use('ggplot')
plt.figure(figsize=(14,5))
trend_df[keyword].plot()
plt.title('Google Trends over time', size = 15)
plt.legend(labels=[keyword], loc='upper right')

# 그래프 파일 저장
# os 모듈의 getcwd() 함수를 이용하여 현재 작업 디렉터리(폴더) 경로를 지정
cwd = os.getcwd()
output_filepath = os.path.join(cwd, 'output', 'google_trend_over_time_%s.png' % keyword)
plt.savefig(output_filepath, dpi=300)
plt.show()
```
