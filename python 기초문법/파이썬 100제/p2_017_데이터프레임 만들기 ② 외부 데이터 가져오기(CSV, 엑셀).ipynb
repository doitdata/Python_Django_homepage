```python
import pandas as pd

# 예제 15에서 지정한 csv 파일명을 경로와 함께 지정
csv_file = './data/bok_statistics_CD.csv'

# read_csv() 함수로 데이터 프레임 변환

# 아무 옵션을 설정 안하면 첫번째 행을 header로 인식되어 데이터프레임의 열 이름 배열로 변환 
df1 = pd.read_csv(csv_file) 
print(df1)
print('\n')

# header = None 옵션 의 경우 모든 행이 데이터 프레임의 행으로 변환, 열이름은 정수 0~2가 자동으로 지정
df2 = pd.read_csv(csv_file, header = None) 
print(df2)
print('\n')

# index_col = 0 옵션의 경우, 첫번째 열이 행 인덱스로 변환 
df3 = pd.read_csv(csv_file, index_col = 0) 
print(df3)
print('\n')

# header = None, index_col = 0  옵션 인 경우, 열이름 정수 0~2가 자동으로 지정되고, 첫번째 열이 행 인덱스 변환 
df4 = pd.read_csv(csv_file, index_col = 0, header = None) 
print(df4)
print('\n')

# 예제 014 에서 다운로드 폴더에 지정한 Excel 파일은 data 폴더에 이동하여 저장
excel_file = './data/report_Key100Stat.xls'

# read_excel() 함수로 데이터 프레임 변환
df5 = pd.read_excel(excel_file)
print(df5)
```
